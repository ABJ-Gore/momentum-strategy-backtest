```python
momentum_data_path= Path("../data/momentum_data")
signals_data_path= Path("../data/signals")

#list of dates(first trading day of every month of 2024) using the first file in the momentum_data folder to get that list
ref_file = momentum_data_path/"AAPL.csv"
ref_df = pd.read_csv(ref_file, parse_dates=["Date"])
ref_df = ref_df[ref_df["Date"].dt.year==2024]


first_days = ref_df["Date"].groupby(ref_df["Date"].dt.month).min()

for target_date in first_days:
    all_data=[]
    for file in momentum_data_path.iterdir():
        df = pd.read_csv(file, parse_dates=["Date"])
        row = df[df["Date"]==target_date]

        if not row.empty:
            row["Symbol"]= file.stem
            all_data.append(row)


    if not all_data:
        print(f"No data for {target_date.date()}")
    
    df_all = pd.concat(all_data).set_index("Symbol")

    df_all["momentum_score"] = (
        0.4 * df_all["ret_12m"] +
        0.3 * df_all["ret_6m"] +
        0.2 * df_all["ret_3m"] +
        0.1 * df_all["ret_1m"]
    )

    df_all = df_all.sort_values("momentum_score", ascending=False)
    df_all["signal"] = "avoid"

    df_all.iloc[:20, df_all.columns.get_loc("signal")] = "long"
    df_all.iloc[-20:, df_all.columns.get_loc("signal")] = "short"

    # Save output
    filename = target_date.strftime("%Y-%m-%d")+".csv"
    df_all.to_csv(signals_data_path / filename)
    print(f" Signals for {target_date.date()} saved.")


  
```
